## Split data

```python
from sklearn.model_selection import train_test_split
```

```python
application_train.shape
```

```python
train, test = train_test_split(application_train, test_size=0.2, random_state=42)
```

```python
train, dev = train_test_split(train, test_size=0.25, random_state=42)
```

## Compare distributions of train/dev/test sets

```python
train['TARGET'].mean()
```

```python
print('Target rates:\n')
print(f"Train target rate: {train['TARGET'].mean()}")
print(f"Dev target rate: {dev['TARGET'].mean()}")
print(f"Test target rate: {test['TARGET'].mean()}")
```

```python
for feature in ['AMT_CREDIT', 'AMT_INCOME_TOTAL', 'AMT_GOODS_PRICE', 'DAYS_BIRTH', 'DAYS_EMPLOYED']:

    f, axes = plt.subplots(1, 3, sharey=True, sharex=True, figsize=(12, 6))

    f.suptitle(feature) 

    axes[0].title.set_text('Train')
    axes[1].title.set_text('Dev')
    axes[2].title.set_text('Test')

    sns.distplot(train[feature], ax=axes[0])
    sns.distplot(dev[feature], ax=axes[1])
    sns.distplot(test[feature], ax=axes[2])
```

```python
for feature in ['AMT_INCOME_TOTAL']:

    f, axes = plt.subplots(1, 3, sharey=True, sharex=True, figsize=(12, 6))

    f.suptitle(feature) 

    axes[0].title.set_text('Train')
    axes[1].title.set_text('Dev')
    axes[2].title.set_text('Test')
    
    sns.boxplot(data=train, y=feature, showfliers=False, ax=axes[0])
    sns.boxplot(data=dev, y=feature, showfliers=False, ax=axes[1])
    sns.boxplot(data=test, y=feature, showfliers=False, ax=axes[2])
```

## Save train/dev/test data

```python
train.to_parquet('data/interim/train.parquet', index=False)
dev.to_parquet('data/interim/dev.parquet', index=False)
test.to_parquet('data/interim/test.parquet', index=False)
```
